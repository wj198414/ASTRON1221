The period right before merging $\delta t$ is about 0.01 second. That tells you the orbital period P of the two merging black holes. When two black holes merge, their orbital distance a is twice the Schwalzschild radius, i.e., 2Rsch, with Rsch = $\frac{2GM}{c^2}$. We also know that orbital velocity v is $\sqrt{\frac{GM}{a}}$. To put it together: $\delta t = 2\pi a / v = 2\pi\frac{4GM}{c^2}/(c/2) = 16\pi GM / c^3$. Rearrange to slove for M, we have $M = \frac{\delta t c^3}{16G\pi}$

```python
delta_t = 0.01 * u.second
M_blackholes = delta_t * ac.c**3 / (16 * np.pi * ac.G)
print(M_blackholes.to(u.solMass))
```

The strain h, which is the y amplitude of the figure above, is $10^{-21}$. The decay of h is inversely proportional to the distance traveled D, and the intital strain at the merger event is proportional to Rsch, so h $\sim$ Rsch / D, and D = Rsch / h

```python
h = 1e-21
Rsch = 2 * ac.G * M_blackholes / 2.0 / ac.c**2 # dividing by 2 because M_blackholes is the total mass of the two black holes
print("Rsch = ", Rsch.to(u.km))
D = Rsch / h
print("Distance of the merger = ", D.to(1e6 * u.parsec))
```
